```python
# Import libraries:
import numpy as np
import pandas as pd
import seaborn as sns  # optional, but recommended
import matplotlib.pyplot as plt 
from sklearn import datasets
```

```python
cancer = datasets.load_breast_cancer()
```

```python
df = pd.DataFrame(cancer.data, columns=cancer.feature_names)
```

```python
# Extract the labels
labels = pd.Series(cancer.target, name='label')

# You can also add the labels to the DataFrame if you want
df['label'] = labels

# Count the number of rows for each label
label_counts = df['label'].value_counts()

print(label_counts)
```

```python
# Define colors for the labels
colors = {0: 'red', 1: 'blue'}

# Create the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df['mean compactness'], df['mean perimeter'], 
            c=df['label'].map(colors), alpha=0.5)
plt.xlabel('Mean Compactness')
plt.ylabel('Mean Perimeter')
plt.title('Scatter Plot of Mean Compactness vs Mean Perimeter')
plt.legend(handles=[plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='red', markersize=10, label='Malignant'),
                    plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='blue', markersize=10, label='Benign')])
plt.show()
```

```python
# Create a DataFrame with the selected feature data
df_selected = df[['mean texture', 'mean perimeter', 'mean compactness', 'mean radius']]

# Add the labels to the DataFrame
df_selected['label'] = pd.Series(cancer.target, name='label')

# Create the pairplot with Seaborn
plt.figure(figsize=(12, 12))
sns.pairplot(df_selected, hue='label', palette={0: 'red', 1: 'blue'}, markers=["o", "s"])
plt.suptitle('Pairwise Scatterplots of Selected Features', size=16)
plt.show()
```
